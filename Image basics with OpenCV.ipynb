```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = cv2.imread('./data/dog.jpg')
type(img)
```

```python
# OpenCV reads channels as blue green red by default
# must use the cvtColor function to fix order of channels
```

```python
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
plt.show()
```

```python
gray_image = cv2.imread("./data/dog.jpg", cv2.IMREAD_GRAYSCALE)
gray_image.shape
```

```python
# Need to specify to mathplotlib that the values should be presented as grayscale
plt.imshow(gray_image, cmap="gray")
plt.show()
```

```python
rgb_image.shape
```

```python
# (width, height)
resizedImage = cv2.resize(rgb_image, (710,440))
print(resizedImage.shape)
plt.imshow(resizedImage)
plt.show()
```

```python
w_ratio = .5
h_ratio = .5
new_img = cv2.resize(rgb_image,(0,0), rgb_image, w_ratio, h_ratio)
print(new_img.shape)
plt.imshow(new_img)
plt.show()
```

```python
# Input values of 1, 0, -1
# 1 will flip the image about the y-axis
# 0 will flip the image about the x-axis
# -1 flips image about the x-axis and y-axis
flippedImage = cv2.flip(rgb_image,1)
plt.imshow(flippedImage)
plt.show()
```

```python

flippedImage = cv2.flip(rgb_image,-1)
plt.imshow(flippedImage)
plt.show()
```

```python
flippedImage = cv2.flip(rgb_image,0)
plt.imshow(flippedImage)
plt.show()
```

```python
# save file
cv2.imwrite("./data/flippedDog.jpg", flippedImage)
```
