```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
pic = Image.open('./data/dog.jpg')
type(pic)
```

```python
picArr = np.asarray(pic)
type(picArr)
```

```python
# (height, width, channels)
picArr.shape
```

```python
plt.imshow(picArr)
plt.show()
```

## Picture channels 

```python
pic = picArr.copy()
```

```python
# Values within channels are from 0 - 255 
# 0 = no color and 255 = Full color
```

```python
# Red channel
plt.imshow(pic[:,:,0], cmap="gray")
plt.show()
```

```python
# Green channel
plt.imshow(pic[:,:,1], cmap="gray")
plt.show()
```

```python
# Blue channel
plt.imshow(pic[:,:,2], cmap="gray")
plt.show()
```

## Isolate channel from picture

```python
pic[:,:,0] = 0
pic[:,:,2] = 0
greenPic = pic
```

```python
plt.imshow(greenPic)
plt.show()
```
